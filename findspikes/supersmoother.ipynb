```python
import os
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from supersmoother import SuperSmoother, LinearSmoother
```

```python
fname = os.path.join('/Users/rbiswas/data/LSST/Bryce_kurucz_model/', 'km01_7000.fits_g40_7000')
```

```python
data = np.loadtxt(fname)
```

```python
y = data[:, 1]
x = data[:, 0]
# Not sure if it will work with no errors. Choose with equal weights 
dy = np.ones(len(y))
```

```python
plt.plot( data[:, 0], data[:, 1])
plt.xscale('log')
```

```python
model = SuperSmoother()
model.fit(x, y, dy)
```

```python
plt.plot(x, model.predict(x), '-k', lw=2., label='smooth')
plt.plot(x, y, color='r', ls='dashed', label='real')
plt.xscale('log')
plt.legend(loc='best')
```

```python
plt.plot(x, model.predict(x)/y)
# plt.xscale('log')
plt.xlim(xmax=100)
```

```python
plt.plot(x, y)
plt.yscale('log')
plt.xlim(xmax=200)
```

This is because all values are identically 0, ie. probably wrong before lambda ~92. These seem to padding

```python
plt.plot(x, y - model.predict(x))
plt.xscale('log')
```

```python
m2 = SuperSmoother()
```

```python
m2.fit(x, y - model.predict(x), dy=1)
```

```python
plt.plot(x, m2.predict(x), 'b')
plt.plot(x, model.predict(x) - y, '-g')
plt.plot(x, m2.predict(x)/y ,'r')
#plt.xlim(xmax=1000)
plt.yscale('log')
plt.xscale('log')
```
