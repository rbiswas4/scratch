```python
import corner
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
ndim, nsamples = 3, 5000
data1 = np.random.randn(ndim *4 * nsamples/5).reshape([4 *nsamples /5, ndim])
data2 = np.random.randn(ndim *4 * nsamples * 16 /5).reshape([4 *nsamples *16 /5, ndim])
```

```python
print np.shape(data1), np.shape(data2)
```

```python
figure = corner.corner(data1, labels=[r"$x$", r"$y$", r"$\log \alpha$",
                                     r"$\Gamma \, [\mathrm{parsec}]$"],
                         truths=[0.0, 0.0, 0.0], hist_kwargs={'normed':True},
                         quantiles=[0.16, 0.5, 0.84],
                         show_titles=True, title_args={"fontsize": 12})
figure.gca().annotate("A Title", xy=(0.5, 1.0), xycoords="figure fraction",
                      xytext=(0, -5), textcoords="offset points",
                      ha="center", va="top")
```

```python
figure2 = corner.corner(data2, labels=[r"$x$", r"$y$", r"$\log \alpha$",
                                     r"$\Gamma \, [\mathrm{parsec}]$"],
                         truths=[0.0, 0.0, 0.0], hist_kwargs={'normed':True},
                         quantiles=[0.16, 0.5, 0.84],
                         show_titles=True, title_args={"fontsize": 12}, fig=figure)
figure2.gca().annotate("A Title", xy=(0.5, 1.0), xycoords="figure fraction",
                      xytext=(0, -5), textcoords="offset points",
                      ha="center", va="top")
```
